```python
def karelerial():
    for i in range(1,6):
        yield i ** 2 # yield anahtar kelimesi generator'un değer üretmesi için kullanılıyor.
generator =  karelerial()
 
print(generator) #
```

```python
iterator = iter(generator)
 
print(next(iterator)) # 1 değeri üretildi
print(next(iterator)) # 4 değeri üretildi 1 değeri tarihe karıştı.
print(next(iterator)) # 9 değeri üretildi 4 değeri tarihe karıştı.
print(next(iterator)) # 16 değeri üretildi 9 değeri tarihe karıştı
print(next(iterator)) # 25 değeri üretildi 16 değeri tarihe karıştı.
print(next(iterator)) # Üretilecek değer kalmadı.
 
```

```python
def carpimtablosu():
    for i in range(1,11):
        for j in range(1,11):
            yield "{} x {} = {}".format(i,j,i*j)
```

```python
for i in carpimtablosu():
    print(i)
 
```

```python
class Kareler():
    
    def __init__(self,maksimum):
        self.maksimum = maksimum
        
        self.sayı = 1
        
    def __iter__(self):
        return self
    def __next__(self):
        
        if (self.sayı <= self.maksimum):
            
            sonuc =  self.sayı ** 2
            self.sayı += 1
            return sonuc
        else:
            self.sayı = 1
            raise StopIteration
    
        
```

```python
kareler = Kareler(5)
```

```python
for i in kareler:
    print(i)
```

```python
def asal_mı(sayı):
    i =  2
    
    while i < sayı:
        if (sayı % i == 0):
            return False
        i += 1
    return True
def asal_generator():
    i =  2
    while True:
        if (asal_mı(i)):
            yield i
        i += 1

for sayı in asal_generator():
    if (sayı > 1000):
        break
    print(sayı)
```
